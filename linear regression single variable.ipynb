```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
data = pd.read_csv('canada_per_capita_income.csv')
data.head()
```

```python
plt.scatter(data = data, x = 'year', y = 'per capita income (US$)', color = 'green', marker = '+')
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.title('price per caipta income through the years')
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(data[['year']], data['per capita income (US$)'])
```

```python
reg.predict([[2020]])
```
